FEATURE ENGINEERING AND DATA MODELLING

```python
from sklearn.model_selection import train_test_split
X = df.drop(["Chance of Admit "], axis =1)
y = df["Chance of Admit "]

print("shape if X: ", X.shape)
print("shape if y: ", y.shape)
```

TRAIN TEST SPLIT

```python
X_train , X_test , y_train , y_test =train_test_split(X, y ,test_size = 0.2, shuffle = True)
print("X_train shape: {}".format(X_train.shape))
print("X_test shape: {}".format(X_test.shape))
print("y_train shape: {}".format(y_train.shape))
print("y_test shape: {}".format(y_test.shape))
```

### FEATURE STANDARDIZATION

```python
from sklearn.preprocessing import StandardScaler
X_train_columns = X_train.columns
std = StandardScaler()
X_train_std=std.fit_transform(X_train)
```

```python
X_train_std
```

```python
X_train = pd.DataFrame(data = X_train_std, columns = X_train_columns)
X_train.head(5)
```

MODEL BUILDING - LINEAR REGRESSION , LASSO , RIDGE

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_squared_error
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
lm = LinearRegression()
lm.fit(X_train.values , y_train)
predictions = lm.predict(std.transform(X_test))
print("RMSE Linear Regression " , np.sqrt(mean_squared_error(y_test,predictions)))
list(zip(X_train.columns , lm.coef_))
```

```python
lm = Lasso()
lm.fit(X_train.values,y_train)
predictions = lm.predict(std.transform(X_test))
print("RMSE Lasso", np.sqrt(mean_squared_error(y_test,predictions)))
list(zip(X_train.columns , lm.coef_))
```

```python
lm = Ridge()
lm.fit(X_train.values,y_train)
predictions = lm.predict(std.transform(X_test))
print("RMSE Ridge", np.sqrt(mean_squared_error(y_test,predictions)))
list(zip(X_train.columns , lm.coef_))
```

MODEL SUMMARY USING STATS MODEL

```python
def build_model(X,y):
  X = sm.add_constant(X) # adding a constant
  lm = sm.OLS(y.values,X).fit() # fitting the model
  print(lm.summary()) # model summary
  return X
```

```python
def check_vif(X):
  vif = pd.DataFrame()
  vif["Features"] = X.columns
  vif["VIF"] = [variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
  vif["VIF"] = round(vif["VIF"],2)
  vif = vif.sort_values(by = "VIF", ascending = False)
  return vif
```

MODEL 1

```python
X_train_new = build_model(X_train, y_train)
```

```python
check_vif(X_train_new)
```

Inference

p-value of SOP seems to be higher than the significance value of 0.05, hence dropping it as it is insignificant in presence of other variables.

```python
X_train_new  = X_train_new.drop(["SOP"],axis =1)
```

MODEL 2

```python
X_train_new = build_model(X_train_new, y_train)
```

p-value of University Rating  seems to be higher than the significance value of 0.05, hence dropping it as it is insignificant in presence of other variables.

```python
X_train_new  = X_train_new.drop(["University Rating"],axis =1)
```
